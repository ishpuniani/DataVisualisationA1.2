```
import pandas as pd
import altair as alt
import numpy as np
from vega_datasets import data

DATA_FILE = '/content/drive/My Drive/Colab Notebooks/DataVisualisation/minard-data.csv'
df = pd.read_csv(DATA_FILE)

text_color = '#fff100'
bg_color = '#000079'
temp_grid_color = '#fff100'
temp_text_color = '#fff100'
temp_line_color = '#ff005d'
map_a_color = '#9600ff'
map_r_color = '#00f5fb'

cities = df[['LONC','LATC','CITY']][:20]

cities_chart = alt.Chart(cities).mark_text(
    font='Cardo',
    fontSize=11,
    fontStyle='italic',
    dx=-3,
    align='left',
    color=text_color
).encode(
    longitude='LONC',
    latitude='LATC',
    text='CITY'
)

temperature = df[['LONT',	'TEMP',	'DAYS',	'MON',	'DAY']][:9]
temperature["label"] = temperature.fillna("").apply(
    axis=1, func=lambda row: "{}° {}".format(row[1], row[3] + ' ' + str(int(row[4])))
)

x_encode = alt.X(
    'LONT', 
    scale=alt.Scale(
        domain=[cities["LONC"].min(), cities["LONC"].max()]
    ),
    axis=None
)

y_encode = alt.Y(
    'TEMP',
    axis=alt.Axis(
        title="Temperature on Retreat", 
        # grid=True, 
        orient='right',
        titleColor=temp_grid_color,
        labelColor=temp_grid_color,
        domainColor=temp_grid_color
    )
)

temperatures_chart = alt.Chart(temperature).mark_line(
    color=temp_line_color
).encode(
    x=x_encode,
    y=y_encode
) + alt.Chart(temperature).mark_text(
    dx=5, 
    dy=20, 
    font='Cardo', 
    fontSize=10,
    color=temp_text_color
).encode(
    x=x_encode,
    y=y_encode,
    text='label'
)
temperatures_chart

temperatures_chart = temperatures_chart.properties(
    height=100
)

def plot_squad_chart(squad, a_color=map_a_color, r_color=map_r_color):
  squad_chart = alt.Chart(squad).mark_trail().encode(
    longitude='LONP',
    latitude='LATP',
    size = alt.Size(
        'SURV', 
        scale = alt.Scale(range=[1,40]),
        legend = None
    ),
    color = alt.Color(
        'DIR',
        scale=alt.Scale(
        domain=['A','R'],
        range=[a_color,r_color],
        )
      )
  )
  return squad_chart

squads = df[['LONP', 'LATP',	'SURV',	'DIR',	'DIV']]

squad_1 = squads[squads['DIV'] == 1]
squad_2 = squads[squads['DIV'] == 2]
squad_3 = squads[squads['DIV'] == 3]

squad_1_temp = squad_1.append(pd.Series([37.6,55.8,100000,'A',1], index=squad_1.columns ), ignore_index=True)
squad_1_chart = plot_squad_chart(squad_1_temp)

squad_2_temp = squad_2.append(pd.Series([28.7,55.5,33000,'A',2], index=squad_2.columns ), ignore_index=True)
squad_2_chart = plot_squad_chart(squad_2_temp)

squad_3_temp = squad_3.append(pd.Series([24.6,55.8,22000,'A',3], index=squad_3.columns ), ignore_index=True)
squad_3_chart = plot_squad_chart(squad_3_temp)

final_squads_chart = squad_3_chart + squad_2_chart + squad_1_chart

squads_text = squads.iloc[::2, :].copy()
squads_text["LONP"] += 0.13 * (squads_text["DIV"])
squads_text["LATP"] += squads_text["DIR"].replace({"A": 0.35, "R": -0.21})

squads_text_chart = alt.Chart(squads_text).mark_text(
    font='Cardo',
    fontSize=7,
    fontStyle='italic',
    angle=280,
    color=text_color,
).encode(
    longitude='LONP:Q',
    latitude='LATP:Q',
    text='SURV'
).project(
    type="mercator"
)

final = final_squads_chart + cities_chart + squads_text_chart

final_chart = alt.vconcat(final, temperatures_chart).configure_view(
    width=900,
    height=400,
    strokeWidth=0
).configure_axis(
    grid=False,
    labelFont="Cardo",
    titleFont="Cardo"
)

final_chart.background=bg_color
final_chart
```
